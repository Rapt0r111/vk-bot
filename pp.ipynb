```python
import vk_api
import numpy as np
import config
import requests
import json
from bs4 import BeautifulSoup
import random
import time
import threading
import datetime
from vk_api.bot_longpoll import VkBotLongPoll, VkBotEventType
from threading import Thread
vk_session = vk_api.VkApi(token=config.token)
vk_session.get_api()
vk = vk_session.get_api()
group_id = config.group_id
print(1)
def get_user_info(user_id):
    return vk.method('users.get', {'user_ids': user_id})

def get_random_id():
    return np.random.randint(0,100000000)

def answer_message(message_id, peer_id, text= '',stick=False,attachment=False):
            query_json = json.dumps({"peer_id": peer_id,"conversation_message_ids":[message_id],"is_reply":True})
            ok = True
            if attachment:
                while not not ok:
                    try:
                        vk.messages.send(peer_id=peer_id, forward = [query_json], random_id=get_random_id(),attachment=f"photo-196348649_{random.randint(457239033, 457270091)}")
                        ok= False
                    except:
                        pass
            elif text !=    '':
                vk.messages.send(peer_id=peer_id, forward = [query_json], message=text, random_id=get_random_id())
            if stick:
                vk.messages.send(peer_id=peer_id,sticker_id=63820, random_id=get_random_id())

def atime():
    while True:
        if int(datetime.datetime.now().hour) == 0 and int(datetime.datetime.now().minute) == 0 and int(datetime.datetime.now().second) == 0:
            vk.method('messages.send', {'peer_id': 2000000003, 'message': '–í—Ä–µ–º—è —Å–ø–∞—Ç—å', 'random_id': get_random_id()})
            time.sleep(1)
        time.sleep(0.33)


answers = ("–≠—Ç–æ –Ω–µ—Å–æ–º–Ω–µ–Ω–Ω–æ", "–ü–µ—Ä—Å–ø–µ–∫—Ç–∏–≤—ã —Ö–æ—Ä–æ—à–∏–µ", "–ú–æ–∂–µ—Ç–µ –Ω–µ —Å–æ–º–Ω–µ–≤–∞—Ç—å—Å—è",\
            "–ú–æ–π –æ—Ç–≤–µ—Ç –Ω–µ—Ç", "–ú–æ–∏ –∏—Å—Ç–æ—á–Ω–∏–∫–∏ –≥–æ–≤–æ—Ä—è—Ç –Ω–µ—Ç", "–¢–æ—á–Ω–æ –¥–∞", " –ù–µ—è—Å–Ω–æ", "–ú–∞–ª–æ —à–∞–Ω—Å–æ–≤",\
            "–î—É–º–∞—é –¥–∞"," –ù–µ—Ç", "–î–∞", "–ë–µ–∑ —Å–æ–º–Ω–µ–Ω–∏–π")
rules = ("""–ü—Ä–∞–≤–∏–ª–∞ –±–∏—Å–µ–¥—ã ü•∞ÔΩ•Ôæü:*–®–ª—ë–øAnime*: ÔΩ•Ôæüü•∞
            –ü—Ä–æ—Å–∏–º —Å–∞–±–ª—é–¥–∞—Ç—å üòô
            1 –ù–∏–∫–∞–≤–æ –Ω–∏ –∞–±–∏–∂–∞—Ç—å, –Ω–µ –∞–±–∑—ã–≤–∞—Ü–∞ —Å–∏–Ω–∞ –∏ –æ–±–∏–¥–Ω–∞ ‚ò∫
            2 –º–æ—Ç–µ—Ä–∏—Ü–∞ ü§¨–º–æ–∂–Ω–∞ –Ω–æ –Ω–µ –º–Ω–æ–≥–∞üòã
            3 —É–≤–æ–∂–∞—Ç—å –∏ —Å–ª—É—à–∞—Ç—å –æ–¥–º–∏–Ω–æ–≤ –±–∏—Å–µ–¥—ã ü§ì
            4 –§–ª—É–¥ –Ω–∏–ª—å–∑—è üö´
            5 –ø—Ä–µ–≥–ª–æ—à–∞—Ç—å —Ç–æ–ª—å–∫–∞ —Å —Ä–æ–∑—Ä–∏—à–µ–Ω–∏—è –æ–¥–º–∏–Ω–æ–≤üò§
            6 –ê—É—Ñ –±–æ–º–±–∏–ª–∞üò±üò±üò±
            7 –±—Ñ—É —Ñ–∏–≥–Ω—èü§¢ü§¢
            8 –¥–æ–ª–≥–∞ –∞—Ñ–∫ = –∫–∏–∫üòà
            9 –ª—é–±–∏—Ç—å –®–ª–µ–ø—É üòçüòç
            10 –Ω–∏ –Ω–∞–∑—ã–≤–∞—Ç—å –®–ª–µ–ø—É –°—Ç–µ–ø–æ–π!!!!!!üò°üò°üò°üò°
            –ó–∞ –Ω–∞—Ä—É—à–µ–Ω–∏—è –ø—Ä–∞–≤–µ–ª –±–∞–Ω (–∫–∏–∫ –∏–∑ –±–∏—Å–µ–¥—ã) ü§¨ü§Øüò°ü•µ""")

def magicball(question):
    if question != '' and question:
        return answers[random.randint(0,len( answers )-1)]
    else:
        return "–°–ø—Ä–æ—Å–∏ –µ—â–µ —Ä–∞–∑"
def rndfact():
    try:
        response = requests.get('https://randstuff.ru/fact/')
        soup = BeautifulSoup(response.content, 'html.parser')
        data = soup.find('table',{'class': 'text'}).text
        return data
    except:
        return '–í –¥–∞–Ω–Ω—ã–π –º–æ–º–µ–Ω—Ç —Ö—É–π –≤–∞–º. –ü–æ–ø—Ä–æ–±—É–π—Ç–µ –ø–æ–∑–∂–µ :)'
def rndask():
    try:
        response = requests.get('https://randstuff.ru/saying/')
        soup = BeautifulSoup(response.content, 'html.parser')
        data = soup.find('table',{'class': 'text'}).text    
        return data.split('‚Äî')[0]
    except:
        return '–í –¥–∞–Ω–Ω—ã–π –º–æ–º–µ–Ω—Ç —Ö—É–π –≤–∞–º. –ü–æ–ø—Ä–æ–±—É–π—Ç–µ –ø–æ–∑–∂–µ :)'

longpoll = VkBotLongPoll(vk_session, group_id)

def main():
    for event in longpoll.listen():
        if event.type == VkBotEventType.MESSAGE_NEW:
            print(event.object.message.get('text'))

            message = event.object.message.get('text').lower()
            peer_id = event.object.message.get('peer_id')
            userID = event.object.message.get('from_id')
            idmess = event.object.message.get('conversation_message_id')

            #give.messages.delete(peer_id=peer_id,conversation_message_ids=2313, delete_for_all=1)
            if userID != 320585774 and '—è –≥–µ–π' in message:
                answer_message(message_id=idmess, peer_id=peer_id, text="–ù–µ –ø—Ä–∞–≤–¥–∞",stick=False, attachment=False)
            elif  message == '!–∞–Ω–∏–º–µ':
                answer_message(message_id=idmess, peer_id=peer_id,stick=False,attachment=True)    
            elif '–∏–¥–∏ –Ω–∞—Ö—É–π' in message or '–ø–æ—à—ë–ª –Ω–∞—Ö—É–π' in message:
                answer_message(message_id=idmess, peer_id=peer_id, text="–°–∞–º –∏–¥–∏ –Ω–∞—Ö—É–π :(",stick=False)
            elif '—Ñ–∞–º–∏–ª–∏—è' in message and '—è–Ω' in message:
                answer_message(message_id=idmess, peer_id=peer_id, text="–ß—É—Ö–∞–ª—ë–Ω–æ–≤",stick=False)
            elif '!—Ñ–∞–∫—Ç' in message:
                vk.method('messages.send', {'peer_id': peer_id, 'message': rndfact(),'random_id': get_random_id()})
            elif '!–º—É–¥—Ä–æ—Å—Ç—å' in message:
                vk.method('messages.send', {'peer_id': peer_id, 'message': rndask(), 'random_id': get_random_id()})
            elif '—Å—Ç–∞—Å' in message:
                answer_message(message_id=idmess, peer_id=peer_id, text="–í–æ–æ–±—â–µ-—Ç–æ –æ–Ω –°–æ–Ω –ö–∏ –•—É–Ω –∏–∑ –°–∞–Ω-–ú—É–Ω–¥–æ–Ω–∞",stick=False)
            elif '!—à–∞—Ä' in message:
                answer_message(message_id=idmess, peer_id=peer_id, text=magicball(question=message[4:].split()),stick=False)
            elif '!–ø—Ä–∞–≤–∏–ª–∞' in message:
                vk.method('messages.send', {'peer_id': peer_id, 'message': rules, 'random_id': get_random_id()})
                
def inpcmd():
    while True:
        command = input()
        if command[:7]=='!–í—ã–≤–µ–¥–∏':
            vk.method('messages.send', {'peer_id': 2000000003, 'message': command[7:], 'random_id': get_random_id()})
    time.sleep(1)

if __name__ == '__main__':
    t1 = Thread(target = atime, args = (),daemon=True)
    t2 = Thread(target = main, args = ())
    t3 = Thread(target = inpcmd, args = ())
    t1.start()
    t2.start()
    t3.start()
```
